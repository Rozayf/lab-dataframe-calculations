```python
def cleanup(x):
    if 'Mega' in x['Name']:
        return 'Mega' + x['Name'].partition('Mega')[2]
    else:
        return x['Name']

pokemon['Name'] = pokemon.apply(cleanup, axis=1)
```

```python
#pokemon['Name']= pokemon['Name'].str.replace('.*Mega', 'Mega')
```

```python
pokemon['A/D Ratio'] = pokemon['Attack']/pokemon['Defense']
```
